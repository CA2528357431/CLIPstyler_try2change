```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def func(i):

    f = open(f"../data/{i}.txt", "r")
    strings = f.readlines()
    li =[[float(x) for x in string.split()] for string in strings]
    return li
```

```python
lis = func("nums")
print(len(lis),len(lis[0]))

li = []
for x in lis:
    li+=x
```

```python
plt.hist(li, bins=64, alpha=0.5, color=["r"]*1, histtype='stepfilled')
```

```python
print(len(lis))
liss = [lis[i] for i in [0,19,39,59,79,99]]
```

```python
colors = ["red", "orange", "yellow", "green", "blue", "purple"]
labels = [str(i*20) for i in range(6)]
```

```python
plt.hist(liss, bins=16, alpha=0.5, color=colors, histtype='bar', label=labels)
plt.legend()
```
