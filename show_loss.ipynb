```python
import matplotlib.pyplot as plt
```

```python
file_ori = open("neo.txt", "r")
loss_li_ori = file_ori.readlines()[0]
loss_li_ori = [float(x) for x in loss_li_ori.split()]
```

```python
file_cmp_inner = open("neo_cmp_inner.txt", "r")
loss_li_cmp_inner = file_cmp_inner.readlines()[0]
loss_li_cmp_inner = [float(x) for x in loss_li_cmp_inner.split()]
```

```python
file_cmp_outer = open("neo_cmp_outer.txt", "r")
loss_li_cmp_outer = file_cmp_outer.readlines()[0]
loss_li_cmp_outer = [float(x) for x in loss_li_cmp_outer.split()]
```

```python
x_ori = [i for i in range(100)]
plt.plot(x_ori,loss_li_ori,color="red",label="ours",marker=".")

x_cmp = [i for i in range(200)]
plt.plot(x_cmp,loss_li_cmp_inner,color="blue",label="old idea",marker=".")

x_cmp = [i for i in range(200)]
plt.plot(x_cmp,loss_li_cmp_outer,color="purple",label="origin",marker=".")


plt.xlabel('iteration')
plt.ylabel('loss')

plt.legend()
```
