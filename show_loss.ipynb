```python
import matplotlib.pyplot as plt
```

```python
file_uu = open("neo0.txt", "r")
loss_li_uu = file_uu.readline()
loss_li_uu = [float(x) for x in loss_li_uu.split()]
cur_times_uu = int(file_uu.readline())
loss_li_uu = [x/cur_times_uu for x in loss_li_uu]
```

```python
file_ori = open("neo.txt", "r")
loss_li_ori = file_ori.readline()
loss_li_ori = [float(x) for x in loss_li_ori.split()]
cur_times_ori = int(file_ori.readline())
loss_li_ori = [x/cur_times_ori for x in loss_li_ori]
```

```python
file_cmp_inner = open("neo_cmp_inner.txt", "r")
loss_li_cmp_inner = file_cmp_inner.readline()
loss_li_cmp_inner = [float(x) for x in loss_li_cmp_inner.split()]
cur_times_cmp_inner = int(file_cmp_inner.readline())
loss_li_cmp_inner = [x/cur_times_cmp_inner for x in loss_li_cmp_inner]
```

```python
file_cmp_outer = open("neo_cmp_outer.txt", "r")
loss_li_cmp_outer = file_cmp_outer.readline()
loss_li_cmp_outer = [float(x) for x in loss_li_cmp_outer.split()]
cur_times_cmp_outer = int(file_cmp_outer.readline())
loss_li_cmp_outer = [x/cur_times_cmp_outer for x in loss_li_cmp_outer]
```

```python
x_uu = [i for i in range(100)]
plt.plot(x_uu,loss_li_uu,color="red",label="ours")

x_ori = [i for i in range(100)]
plt.plot(x_ori,loss_li_ori,color="green",label="old model")

x_cmp = [i for i in range(200)]
plt.plot(x_cmp,loss_li_cmp_inner,color="blue",label="old idea")

x_cmp = [i for i in range(200)]
plt.plot(x_cmp,loss_li_cmp_outer,color="purple",label="origin")


plt.xlabel('iteration')
plt.ylabel('loss')

plt.legend()
```
