```python
import matplotlib.pyplot as plt
```

```python
file_ori = open("neo.txt", "r")
loss_li_ori = file_ori.readlines()[0]
loss_li_ori = [float(x) for x in loss_li_ori.split()]
```

```python
file_cmp = open("neo_cmp_inner.txt", "r")
loss_li_cmp = file_cmp.readlines()[0]
loss_li_cmp = [float(x) for x in loss_li_cmp.split()]
```

```python
x_ori = [i for i in range(100)]
plt.plot(x_ori,loss_li_ori,color="red",label="ours",marker=".")

x_cmp = [i for i in range(200)]
plt.plot(x_cmp,loss_li_cmp,color="blue",label="origin",marker=".")

plt.xlabel('iteration')
plt.ylabel('loss')

plt.legend()
```
